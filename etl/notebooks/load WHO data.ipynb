```python
sym = pd.read_csv('../source/ddf--synonyms--country.csv')
```

```python
sym = sym.set_index('synonym')['country'].to_dict()
```

```python
data['cause of death'].unique()
```

```python
m = {'Road traffic accidents': 'traffic', 'Homicide': 'homicide', 'Self-inflicted injuries': 'suicide'}
```

```python
for n in data.name.unique():
    if n not in sym:
        print(n)
```

```python
sym['United kingdom'] = 'gbr'
```

```python
data['cause'] = data['cause'].map(m)
```

```python
data['geo'] = data['name'].map(sym)
```

```python
data = data.rename(columns={'cause of death': 'cause'})
```

```python
data = data.drop(['ICD', 'country', 'name'], axis=1)
```

```python
data = data.set_index(['geo', 'year', 'cause', 'sex'])
```
